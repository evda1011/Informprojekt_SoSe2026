```python
import requests
import pandas as pd
from pathlib import Path

def load_destatis_rauchverhalten():
    url = "https://www.destatis.de/DE/Themen/Gesellschaft-Umwelt/Gesundheit/Gesundheitszustand-Relevantes-Verhalten/Tabellen/liste-rauchverhalten.html"
    
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
    
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    
    # Загружаем таблицы
    dfs = pd.read_html(response.text, flavor='html5lib')
    print(f"✅ Найдено таблиц: {len(dfs)}\n")
    
    df = dfs[0].copy()
    
    # === Улучшенная очистка колонок ===
    df.columns = [
        str(col).strip()
        .replace('\n', ' ')
        .replace('  ', ' ')
        .replace(' / ', '/') 
        for col in df.columns
    ]
    
    print("📋 Колонки после очистки:")
    print(df.columns.tolist())
    print("\n📊 Первые 10 строк:")
    print(df.head(10).to_string())
    
    # Сохранение
    output_dir = Path("data_clean")
    output_dir.mkdir(exist_ok=True)
    
    output_path = output_dir / "rauchverhalten_destatis_2025.csv"
    df.to_csv(output_path, index=False, encoding='utf-8')
    
    print(f"\n✅ Данные успешно сохранены: {output_path}")
    return df


# === Запуск ===
if __name__ == "__main__":
    df = load_destatis_rauchverhalten()
```
