```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
from raucher_daten_jugend_1979_2023_ import daten_tabelle

print(" Jahr    Nie-Raucher    Raucher")
print("─────┼───────────────┼─────────")
for zeile in daten_tabelle:
    print(f"{zeile[0]:4d} │ {zeile[1]:13.2f} │ {zeile[2]:7.2f}")
```

```python
def analysiere_und_visualisiere_raucher():
    """Einfache Funktion: DataFrame + Berechnung des Verhältnisses + 2 Grafiken"""
   
    # 1. Umwandlung der Daten aus der Tabelle in ein übersichtliches DataFrame
    df = pd.DataFrame(daten_tabelle, columns=['Jahr', 'Nichtraucher (%)', 'Raucher (%)'])
   
    # 2. Berechnung des Verhältnisses (Raucher / Nichtraucher)
    df['Verhältnis'] = df['Raucher (%)'] / df['Nichtraucher (%)']
   
    # 3. Speichern als CSV (praktisches Format für das weitere Projekt)
    df.to_csv('clean_data.csv', index=False)
    print("✅ Daten wurden umformatiert und in clean_data.csv gespeichert")
   
    # 4. Ausgabe der Tabelle mit Verhältnis
    print("\n📊 Tabelle mit Verhältnis:")
    print(df.to_string(index=False))
   
    # 5. Einstellungen für die Grafiken
    sns.set_style("whitegrid") # Seaborn-Stil für ansprechende Grafiken
    plt.rcParams['figure.figsize'] = (12, 6)
   
    # Grafik 1: Entwicklung der Anteile
    plt.figure() # Neue Grafik für die Anteile der Raucher und Nichtraucher
    plt.plot(df['Jahr'], df['Nichtraucher (%)'], marker='o', label='Nichtraucher (%)') # Linienplot für Nichtraucher
    plt.plot(df['Jahr'], df['Raucher (%)'], marker='o', label='Raucher (%)') # Linienplot für Raucher
    plt.title('Entwicklung von Rauchern und Nichtrauchern (1979–2023)') # Titel der Grafik
    plt.xlabel('Jahr') # Beschriftung der x-Achse
    plt.ylabel('Prozent (%)') # Beschriftung der y-Achse
    plt.legend() # Legende hinzufügen
    plt.grid(True) # Gitterlinien hinzufügen
    plt.tight_layout() # Layout anpassen
    plt.savefig('raucher_trend.png', dpi=300) # Grafik speichern
    plt.show() # Grafik anzeigen
   
    # Grafik 2: Verhältnis Raucher zu Nichtrauchern
    plt.figure() # Neue Grafik für das Verhältnis
    plt.plot(df['Jahr'], df['Verhältnis'], marker='o', color='red', label='Verhältnis') # Rotes Linienplot für das Verhältnis
    plt.title('Verhältnis Raucher zu Nichtrauchern (1979–2023)')
    plt.xlabel('Jahr') 
    plt.ylabel('Verhältnis') 
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('raucher_verhaeltnis.png', dpi=300)
    plt.show() # Grafik anzeigen
   
    print("✅ Grafiken wurden gespeichert: raucher_trend.png und raucher_verhaeltnis.png")
```

```python
analysiere_und_visualisiere_raucher()   # ← einfach diese Zeile ausführen
```
